```python
import ROOT
from IPython.display import IFrame
```

```python
ROOT.gStyle.SetOptStat(0)
can = ROOT.TCanvas("can", "canvas", 800, 800)

fa = ROOT.TFile("nano2pico/FRUL.root")
fb = ROOT.TFile("nano2pico/FRReco.root")

h1a = fa.Get("h_den")
h2a = fa.Get("h_num")
h2a.Divide(h1a)
h2a.SetMaximum(0.5)
h2a.SetMinimum(0.0)
h2a.SetTitle("Photon Fake Rate")
h2a.GetXaxis().SetTitle("Photon #eta")
h2a.SetLineColor(2)

h1b = fb.Get("h_den")
h2b = fb.Get("h_num")
h2b.Divide(h1b)
h2b.SetLineColor(3)

legend = ROOT.TLegend(0.52, 0.7, 0.9, 0.9)
legend.AddEntry(h2a, "Ultra Legacy", "l")
legend.AddEntry(h2b, "ReReco", "l")

h2a.Draw()
h2b.Draw("SAME")
legend.Draw("SAME")
can.Draw()
```
